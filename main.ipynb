```python
import cv2
import numpy as np
```

```python
screen_width = 1920  
screen_height = 1080 

cv2.namedWindow('RGB Tarama', cv2.WND_PROP_FULLSCREEN)
cv2.setWindowProperty('RGB Tarama', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

for i in range(256):
    current_rgb = np.array([i, i, i])
    img = np.zeros((screen_height, screen_width, 3), dtype=np.uint8)
    img[:] = current_rgb
    cv2.putText(img, f'RGB: ({i},{i},{i})', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 255, 255), 3)
    cv2.imshow('RGB Tarama', img)
    screenshot_filename = f'screenshot_rgb_{i}.png'
    cv2.imwrite(screenshot_filename, img)
    print(f"Ekran görüntüsü kaydedildi: {screenshot_filename}")
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
